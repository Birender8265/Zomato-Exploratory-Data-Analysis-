```python
country = pd.read_excel("Country-Code.xlsx")
country.head()
```

```python
final_data = data.merge(country,how="left",on="Country Code")
```

```python
final_data.head(2)
```

```python
# To cheack Data Type 
final_data.dtypes
```

```python
final_data.columns
```

```python
# # Renaming columns according to snake casing for better readability and documentation
final_data.columns = final_data.columns.str.lower()
final_data.columns = final_data.columns.str.replace(" " , "_")
final_data.columns
```

```python
country_name = final_data["country"].value_counts().index
country_value = final_data["country"].value_counts().values
```

```python
# pie chart -- Top 3 countries that uses zomato
plt.pie(country_value[:3],labels=country_name[:3],autopct="%1.2f%%")
plt.show()
```

 Observation :
* Zomato maximum records or transaction are from India
* After that USA and Then United Kingdoms

```python
ratings = final_data.groupby(['aggregate_rating','rating_color','rating_text']).size().reset_index().rename(columns={0:'rating_count'})
```

```python
ratings
```

Observation : 
* When rating is between 4.5 to 4.9 ---> Excellent
* When rating are between 4.0 to 4.4 ---> Very good
* When rating are between 3.5 to 3.9 ---> Good
* When rating are between 3.0 to 3.4 ---> Average
* When rating are between 2.5 to 2.9 ---> Average
* When rating are between 2.0 to 2.4 ---> Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] =(12,6)
sns.barplot(x='aggregate_rating',y='rating_count',data=ratings)
```

```python
sns.barplot(x='aggregate_rating',y='rating_count',hue='rating_color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Observation :
* Not rated count is very high 
* Maximum number of rating are between 2.5 to 3.4

```python
# count plot on frequency of rating color 
sns.countplot(x='rating_color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
# Find the countries name that has given 0 rating
zero_rating = final_data[(final_data['aggregate_rating'] == 0) &(final_data['rating_color'] == 'White')]
zero_rating['country'].value_counts()
```

Observation :
* Maximum number of 0 rating are from Indian customers

```python
# Find out Which currency is used by which country
final_data[['country','currency']].groupby('country')['currency'].value_counts().reset_index()
```

```python
# Which countries do have onine delivery option
final_data[final_data['has_online_delivery'] == 'Yes']['country'].value_counts()
```

```python
final_data[['has_online_delivery','country']].groupby(['has_online_delivery','country']).size().reset_index()
```

Observation :
* Online deliveries are available in India and UAE
